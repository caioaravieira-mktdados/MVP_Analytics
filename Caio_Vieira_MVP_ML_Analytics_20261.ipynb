# 1. Definição do problema

## 1.1 Descrição do problema

### **Contexto**

A retenção de clientes é um dos principais desafios enfrentados por instituições financeiras. A perda de clientes, conhecida como *churn*, impacta diretamente a receita, reduz o valor do ciclo de vida do cliente (*Customer Lifetime Value* – CLV) e aumenta os custos relacionados à aquisição de novos consumidores.

Com o crescimento do volume de dados disponíveis, instituições financeiras passaram a utilizar técnicas de Ciência de Dados e Machine Learning para identificar padrões de comportamento capazes de antecipar o risco de evasão de clientes. Dessa forma, é possível direcionar campanhas de retenção, personalizar ofertas e otimizar investimentos em relacionamento.

Neste projeto, serão utilizadas informações demográficas, financeiras e comportamentais dos clientes, como idade, pontuação de crédito, saldo em conta, número de produtos contratados, tempo de relacionamento, nível de atividade, índice de satisfação e histórico de reclamações, com o objetivo de identificar padrões associados ao churn.

### **Problema**

Este projeto tem como objetivo desenvolver um modelo de classificação supervisionada capaz de prever se um cliente irá permanecer ou encerrar seu relacionamento com a instituição financeira.

A variável-alvo utilizada será **Exited**, onde:

* **0** → Cliente permaneceu no banco;
* **1** → Cliente deixou a instituição.

### **Decisão que o modelo apoia**

O modelo tem como finalidade apoiar decisões relacionadas à retenção de clientes. A partir da probabilidade de churn estimada para cada cliente, a instituição pode identificar indivíduos com maior risco de evasão e priorizar ações como campanhas de relacionamento, ofertas personalizadas e programas de fidelização.

### **Usuários da solução**

Os principais interessados nesta solução seriam:

* Equipes de CRM;
* Equipes de Marketing;
* Gestores de Relacionamento com Clientes;
* Analistas de Ciência de Dados;
* Gestores de Negócio responsáveis pela estratégia de retenção.

### **Relevância do problema**

A retenção de clientes representa uma vantagem competitiva para instituições financeiras, pois manter um cliente costuma ser significativamente mais econômico do que conquistar um novo.

Assim, modelos preditivos de churn contribuem para decisões mais assertivas, redução da evasão de clientes, aumento da fidelização e melhor alocação dos recursos destinados às estratégias de relacionamento.


### **Premissas**

Para o desenvolvimento deste projeto foram consideradas as seguintes premissas:

* O dataset representa adequadamente o comportamento dos clientes da instituição financeira.
* As variáveis demográficas, financeiras e comportamentais disponíveis contêm informações relevantes para a previsão de churn.
* Não há vazamento de informações futuras para a variável-alvo durante o processo de modelagem.
* O problema pode ser tratado como uma classificação binária supervisionada.
* Os resultados obtidos representam um MVP e poderão ser aprimorados com dados adicionais e técnicas mais avançadas.


## 1.2 Objetivo do MVP

O objetivo deste MVP é construir, treinar e comparar diferentes modelos de Machine Learning capazes de prever o churn de clientes bancários, avaliando seu desempenho por meio de métricas apropriadas para problemas de classificação. Ao final, será selecionado o modelo com melhor capacidade preditiva e discutidas suas possíveis aplicações em um contexto de negócio.


## 1.3 Tipo de problema

### Tipo do Problema

**Tipo escolhido:** Classificação Supervisionada Binária.

**Justificativa:**

O objetivo deste projeto é prever se um cliente permanecerá ou encerrará seu relacionamento com a instituição financeira, utilizando informações demográficas, financeiras e comportamentais, como idade, pontuação de crédito, saldo em conta, número de produtos contratados, nível de atividade e índice de satisfação.. A variável-alvo (**Exited**) possui apenas duas categorias possíveis: **0** (cliente permaneceu) e **1** (cliente realizou churn).

Como o conjunto de dados contém exemplos históricos rotulados, nos quais o resultado desejado já é conhecido, o problema se caracteriza como um caso de **aprendizado supervisionado**. Além disso, por existirem apenas duas classes possíveis para a variável-alvo, trata-se especificamente de uma **classificação binária**.

O modelo desenvolvido será capaz de estimar a probabilidade de um cliente pertencer à classe "churn", fornecendo suporte às equipes de CRM e Marketing na identificação de clientes com maior risco de evasão e permitindo a adoção de estratégias preventivas de retenção.

> **Comentário:** A escolha pela classificação supervisionada binária é adequada porque o objetivo não é prever um valor numérico contínuo, formar agrupamentos de clientes ou estimar eventos futuros em uma sequência temporal, mas sim classificar cada cliente em uma das duas categorias previamente definidas pela variável-alvo.

**Restrições práticas**

Além da qualidade preditiva, o modelo deverá apresentar tempo de treinamento compatível com um ambiente computacional padrão e manter um nível adequado de interpretabilidade, permitindo compreender quais características dos clientes exercem maior influência sobre a previsão de churn.

> **Comentário:** O sucesso deste MVP não será determinado apenas por uma alta acurácia. A solução será considerada satisfatória caso apresente desempenho consistente em dados não utilizados durante o treinamento, supere o modelo baseline e produza resultados interpretáveis, que possam apoiar decisões de negócio relacionadas à retenção de clientes.

## 6. Pré-processamento e pipeline

Após a definição das variáveis utilizadas na modelagem, foi construída uma etapa de pré-processamento para garantir que os dados estivessem no formato adequado para o treinamento dos algoritmos de Machine Learning.

Como o conjunto de dados possui atributos numéricos e categóricos, foram aplicadas transformações específicas para cada tipo de variável. Todas as etapas foram organizadas em um Pipeline, garantindo que os mesmos tratamentos realizados durante o treinamento fossem automaticamente aplicados aos dados de teste, evitando vazamento de informação (data leakage).

As principais etapas do pré-processamento foram:

- imputação de valores ausentes utilizando a mediana (variáveis numéricas) e a moda (variáveis categóricas);
- padronização das variáveis numéricas por meio do StandardScaler;
- codificação das variáveis categóricas utilizando One-Hot Encoding;
- integração de todas as transformações em um único Pipeline do Scikit-Learn.


```python
# =====================================
# Separação das variáveis por tipo
# =====================================

categorical_features = X.select_dtypes(include="object").columns.tolist()
numerical_features = X.select_dtypes(exclude="object").columns.tolist()

print("Variáveis numéricas:")
print(numerical_features)

print("\nVariáveis categóricas:")
print(categorical_features)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler

# Pipeline numérico
numeric_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler())
    ]
)

# Pipeline categórico
categorical_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("encoder", OneHotEncoder(handle_unknown="ignore"))
    ]
)

# Pré-processador completo
preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numerical_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)

print(preprocessor)
```

```python
# Aplicação do pré-processamento

X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)

print("Treino:", X_train_processed.shape)
print("Teste :", X_test_processed.shape)
```


## **6.1 Decisões de pré-processamento**

Durante a etapa de pré-processamento foram adotadas decisões com o objetivo de garantir que os dados estivessem adequadamente preparados para o treinamento dos modelos e que todo o fluxo pudesse ser reproduzido sem risco de vazamento de informação.

Embora o conjunto de dados não apresentasse valores ausentes, foi mantida uma etapa de imputação no Pipeline utilizando a mediana para variáveis numéricas e a moda para variáveis categóricas. Essa decisão torna a solução mais robusta caso novos dados apresentem valores faltantes.

As variáveis numéricas foram padronizadas por meio do **StandardScaler**, reduzindo diferenças de escala entre os atributos e favorecendo algoritmos sensíveis à magnitude das variáveis, como a Regressão Logística.

As variáveis categóricas (**Geography**, **Gender** e **Card Type**) foram transformadas utilizando **One-Hot Encoding**, permitindo que fossem utilizadas pelos algoritmos sem criar uma relação ordinal artificial entre as categorias.

Também foram removidas as colunas **RowNumber**, **CustomerId** e **Surname**, por representarem apenas identificadores dos registros e não contribuírem para a capacidade preditiva dos modelos. Além disso, a variável **Complain** foi excluída da modelagem por apresentar forte associação com a variável-alvo (**Exited**), caracterizando um potencial caso de *data leakage*.

Por fim, todas as etapas de transformação foram organizadas em um **Pipeline** do Scikit-Learn, garantindo que o mesmo fluxo de pré-processamento aplicado aos dados de treinamento fosse utilizado também no conjunto de teste, evitando inconsistências durante a avaliação dos modelos.

# 8. Treinamento e avaliação inicial

Nesta etapa foram treinados os modelos definidos anteriormente utilizando o conjunto de treinamento preparado durante o pré-processamento.

O objetivo é comparar o desempenho inicial dos algoritmos antes da realização de qualquer ajuste fino de hiperparâmetros, permitindo identificar qual abordagem apresenta maior potencial para o problema de previsão de churn.

Foram avaliados três modelos de classificação:

- Regressão Logística (baseline);
- Random Forest;
- XGBoost.

A comparação foi realizada utilizando métricas de classificação, incluindo Accuracy, Precision, Recall, F1-Score e ROC-AUC, além da análise das matrizes de confusão e curvas ROC.

```python
# ==========================================
# Importação das bibliotecas
# ==========================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler

from sklearn.impute import SimpleImputer

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

from xgboost import XGBClassifier

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    confusion_matrix,
    ConfusionMatrixDisplay,
    RocCurveDisplay
)
```

```python
# ==========================================
# Separação das variáveis
# ==========================================

target = "Exited"

X = df.drop(columns=[target])
y = df[target]

# Removendo colunas que não serão utilizadas
X = X.drop(columns=[
    "RowNumber",
    "CustomerId",
    "Surname",
    "Complain"      # removida por possível data leakage
])

# Variáveis numéricas
numeric_features = X.select_dtypes(include=["int64","float64"]).columns.tolist()

# Variáveis categóricas
categorical_features = X.select_dtypes(include=["object"]).columns.tolist()

print("Numéricas:")
print(numeric_features)

print()

print("Categóricas:")
print(categorical_features)
```

```python
from sklearn.model_selection import train_test_split

# Divisão treino e teste
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Treino:", X_train.shape)
print("Teste :", X_test.shape)
```

```python
# =====================================
# Pipeline de pré-processamento
# =====================================

numeric_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="median")),
        ("scaler", StandardScaler())
    ]
)

categorical_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("onehot", OneHotEncoder(handle_unknown="ignore"))
    ]
)

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)
```

```python
# =====================================
# Modelos candidatos
# =====================================

models = {
    "Regressão Logística": LogisticRegression(max_iter=1000),
    "Random Forest": RandomForestClassifier(
        n_estimators=200,
        random_state=42
    ),
    "XGBoost": XGBClassifier(
        random_state=42,
        eval_metric="logloss"
    )
}
```

```python
# =====================================
# Treinamento e avaliação
# =====================================

results = []

trained_models = {}

for name, model in models.items():

    pipe = Pipeline([
        ("preprocessor", preprocessor),
        ("model", model)
    ])

    pipe.fit(X_train, y_train)

    trained_models[name] = pipe

    pred = pipe.predict(X_test)
    prob = pipe.predict_proba(X_test)[:,1]

    results.append({

        "Modelo": name,
        "Accuracy": accuracy_score(y_test, pred),
        "Precision": precision_score(y_test, pred),
        "Recall": recall_score(y_test, pred),
        "F1": f1_score(y_test, pred),
        "ROC AUC": roc_auc_score(y_test, prob)

    })
```

```python
resultados = (
    pd.DataFrame(results)
      .sort_values("ROC AUC", ascending=False)
      .reset_index(drop=True)
)

resultados
```

## 8.1 Análise dos resultados iniciais

Nesta etapa foram comparados três algoritmos de classificação: Regressão Logística, Random Forest e XGBoost.

Os resultados demonstram que os modelos baseados em árvores apresentaram desempenho superior ao baseline. O Random Forest obteve o melhor resultado geral, alcançando Accuracy de 87,05%, F1-Score de 0,598 e ROC-AUC de 0,862, seguido pelo XGBoost. Não foram observados indícios de underfitting nos modelos baseados em árvores, que conseguiram capturar melhor os padrões presentes nos dados. Além disso, o tempo de treinamento foi baixo para todos os algoritmos, tornando sua aplicação viável neste conjunto de dados.

De forma geral, os resultados estão alinhados com a análise exploratória realizada anteriormente e indicam que o Random Forest será utilizado como referência para a etapa de otimização dos hiperparâmetros.


# 9. Validação e otimização de hiperparâmetros

Após a avaliação inicial dos modelos candidatos, foi realizada uma busca simples por hiperparâmetros utilizando o algoritmo Random Forest.

O objetivo desta etapa é verificar se pequenos ajustes nos parâmetros do modelo podem melhorar sua capacidade preditiva sem aumentar significativamente a complexidade do projeto.

Para isso, foi utilizada a técnica de Grid Search com validação cruzada de cinco partições (5-Fold Cross Validation), utilizando como métrica principal o ROC AUC.

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

pipe_rf = Pipeline([
    ("preprocessor", preprocessor),
    ("model", RandomForestClassifier(random_state=42))
])

param_grid = {
    "model__n_estimators": [100, 200],
    "model__max_depth": [5, 10, None],
    "model__min_samples_split": [2, 5]
}

grid = GridSearchCV(
    estimator=pipe_rf,
    param_grid=param_grid,
    cv=5,
    scoring="roc_auc",
    n_jobs=-1
)

grid.fit(X_train, y_train)

print("Melhores parâmetros:")
print(grid.best_params_)

print("\nMelhor ROC AUC:")
print(round(grid.best_score_,4))
```

```python
best_rf = grid.best_estimator_

pred = best_rf.predict(X_test)
prob = best_rf.predict_proba(X_test)[:,1]

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

print("Accuracy :", round(accuracy_score(y_test,pred),4))
print("Precision:", round(precision_score(y_test,pred),4))
print("Recall   :", round(recall_score(y_test,pred),4))
print("F1 Score :", round(f1_score(y_test,pred),4))
print("ROC AUC  :", round(roc_auc_score(y_test,prob),4))
```

## 9.1 Discussão da otimização

Foi realizada uma otimização dos hiperparâmetros do modelo Random Forest utilizando Grid Search com validação cruzada de 5 folds.

A busca avaliou diferentes combinações para os parâmetros `n_estimators`, `max_depth` e `min_samples_split`, utilizando a métrica ROC AUC como critério de seleção.

A melhor configuração encontrada foi:

- n_estimators = 200
- max_depth = 10
- min_samples_split = 5

Essa configuração apresentou ROC AUC médio de **0,8552** durante a validação cruzada.

Após a aplicação do modelo otimizado no conjunto de teste, obteve-se:

- Accuracy: **0,8730**
- Precision: **0,8500**
- Recall: **0,4583**
- F1-Score: **0,5955**
- ROC AUC: **0,8675**

Comparando com o modelo inicial, observou-se uma pequena melhora no ROC AUC e na precisão, indicando maior capacidade de distinguir clientes com maior probabilidade de churn. Houve uma leve redução no recall, caracterizando um trade-off esperado entre precisão e sensibilidade.

# 10. Avaliação final no conjunto de teste

Após a etapa de otimização de hiperparâmetros, o modelo Random Forest foi selecionado como modelo final e avaliado no conjunto de teste, composto por dados que não participaram do treinamento nem da validação cruzada.

Essa avaliação representa uma estimativa mais realista da capacidade de generalização do modelo em um cenário de aplicação prática.


```python
from sklearn.metrics import (
    classification_report,
    ConfusionMatrixDisplay
)
import matplotlib.pyplot as plt

print("Modelo final: Random Forest otimizado\n")

y_pred = best_rf.predict(X_test)

print(classification_report(y_test, y_pred))

ConfusionMatrixDisplay.from_estimator(
    best_rf,
    X_test,
    y_test,
    cmap="Blues"
)

plt.title("Matriz de Confusão - Modelo Final")
plt.show()
```

A matriz de confusão mostra que o modelo classificou corretamente **1.559 clientes que permaneceram** e **187 clientes que realizaram churn**. Foram observados **33 falsos positivos** e **221 falsos negativos**, indicando que o modelo é bastante preciso quando prevê churn, porém ainda deixa de identificar parte dos clientes que efetivamente cancelam seus serviços.

O Classification Report confirma esse comportamento, apresentando elevada precisão para as previsões positivas e desempenho consistente nas demais métricas. Apesar do recall moderado, o modelo alcançou um bom equilíbrio entre precisão e capacidade discriminatória, evidenciado pelo **ROC AUC de aproximadamente 0,87**.

Considerando as métricas obtidas, o **Random Forest otimizado** apresentou o melhor equilíbrio entre Accuracy, Precision, F1-Score e ROC AUC, sendo escolhido como modelo final para este problema de previsão de churn.

Dessa forma, o modelo deve ser utilizado como ferramenta de apoio à tomada de decisão e priorização de clientes com maior risco de evasão, não substituindo a análise de negócio nem sendo empregado como único critério para decisões estratégicas.

# 12. Boas práticas e rastreabilidade

Esta seção registra as principais decisões tomadas durante o desenvolvimento do projeto, permitindo rastrear as escolhas realizadas em cada etapa e facilitar a reprodutibilidade dos experimentos.

**Resumo das boas práticas adotadas**

Seed utilizada: random_state = 42.

Principais decisões de pré-processamento: remoção de variáveis identificadoras, imputação de valores ausentes, padronização de atributos numéricos e One-Hot Encoding das variáveis categóricas.

Modelos testados: Regressão Logística, Random Forest e XGBoost.

Hiperparâmetros otimizados: número de árvores (n_estimators), profundidade máxima (max_depth) e número mínimo de amostras para divisão (min_samples_split) do Random Forest.

Tempo aproximado de treinamento: inferior a alguns minutos em ambiente Google Colab.

Recursos computacionais utilizados: Python 3, Scikit-learn, XGBoost, Pandas e Google Colab.

Limitações conhecidas: dataset desbalanceado e ausência de variáveis comportamentais mais detalhadas, que poderiam aumentar o poder preditivo.

Abordagens descartadas: a Regressão Logística foi mantida apenas como baseline por apresentar desempenho inferior aos modelos baseados em árvores.

**Registro de decisões:**

| Decisão                                                              | Justificativa                                                    | Impacto esperado                                                             |
| -------------------------------------------------------------------- | ---------------------------------------------------------------- | ---------------------------------------------------------------------------- |
| Utilizar `random_state = 42`                                         | Garantir a reprodutibilidade dos experimentos.                   | Permitir que os resultados sejam reproduzidos em diferentes execuções.       |

| Utilizar divisão Holdout (80% treino / 20% teste) com estratificação | Preservar a proporção das classes de churn.                      | Obter uma avaliação mais confiável do desempenho dos modelos.                |
| Utilizar ROC AUC como métrica principal                              | O problema apresenta desbalanceamento moderado entre as classes. | Avaliar melhor a capacidade discriminatória dos modelos.                     |
| Testar Regressão Logística, Random Forest e XGBoost                  | Comparar modelos lineares e baseados em árvores.                 | Identificar a abordagem com melhor desempenho para o problema.               |
